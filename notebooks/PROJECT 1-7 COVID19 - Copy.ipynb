```python
#ACTIVITY 3: Visualize the total COVID-19 cases and total deaths by continent using bar plots.

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('owid-covid-data.csv')

# Group by continent and sum the total cases and deaths
continent_data = df.groupby('continent')[['total_cases', 'total_deaths']].sum().reset_index()
```

```python
# Plot total cases by continent
plt.figure(figsize=(10, 5))
plt.bar(continent_data['continent'], continent_data['total_cases'], color='blue')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.title('Total COVID-19 Cases by Continent')
plt.show()

# Plot total deaths by continent
plt.figure(figsize=(10, 5))
plt.bar(continent_data['continent'], continent_data['total_deaths'], color='red')
plt.xlabel('Continent')
plt.ylabel('Total Deaths')
plt.title('Total COVID-19 Deaths by Continent')
plt.show()
```

```python
#Explore the monthly trend of new COVID-19 cases using a line plot.

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('owid-covid-data.csv')

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract month and year from the 'date' column
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# Group by year and month, then sum the new cases
monthly_cases = df.groupby(['year', 'month'])['new_cases'].sum().reset_index()

# Create a pivot table for better visualization
monthly_cases_pivot = monthly_cases.pivot(index='month', columns='year', values='new_cases')

# Plotting the line plot
monthly_cases_pivot.plot(kind='line', figsize=(12, 8))
plt.xlabel('Month')
plt.ylabel('New Cases')
plt.title('Monthly Trend of New COVID-19 Cases')
plt.legend(title='Year')
plt.show()
```

```python
#Investigate the correlation between total cases and total deaths using a heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv')
selected_cols = ['total_cases', 'total_deaths']
df_selected = df[selected_cols]
df_selected = df_selected.dropna()
correlation_matrix = df_selected.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Total Cases and Total Deaths')
plt.show()
```

```python
#ACTIVITY 4 Create a grouped bar chart to visualize new cases by continent and month.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('owid-covid-data.csv')

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract year and month from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Group by continent and month, then sum the new cases
grouped_df = df.groupby(['continent', 'month'])['new_cases'].sum().reset_index()

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(data=grouped_df, x='month', y='new_cases', hue='continent')
plt.title('New COVID-19 Cases by Continent and Month')
plt.xlabel('Month')
plt.ylabel('New Cases')
plt.legend(title='Continent')
plt.show()
```

```python
#Calculate and visualise the daily average of new cases and deaths globally. 

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('owid-covid-data.csv')
data['date'] = pd.to_datetime(data['date'])
global_daily_avg = data.groupby('date')[['new_cases', 'new_deaths']].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(global_daily_avg['date'], global_daily_avg['new_cases'], label='Daily Average New Cases', color='blue')
plt.xlabel('Date')
plt.ylabel('Daily Average New Cases')
plt.title('Daily Average New COVID-19 Cases Globally')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(global_daily_avg['date'], global_daily_avg['new_deaths'], label='Daily Average New Deaths', color='red')
plt.xlabel('Date')
plt.ylabel('Daily Average New Deaths')
plt.title('Daily Average New COVID-19 Deaths Globally')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Analyze the total tests and positive rate over time globally. 

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('owid-covid-data.csv')

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Group by date
df_grouped = df.groupby('date').agg({'total_tests': 'sum', 'positive_rate': 'mean'}).reset_index()

# Plotting
plt.figure(figsize=(14, 7))

# Plot total tests
plt.subplot(2, 1, 1)
plt.plot(df_grouped['date'], df_grouped['total_tests'], label='Total Tests')
plt.title('Total Tests Over Time')
plt.xlabel('Date')
plt.ylabel('Total Tests')
plt.legend()

# Plot positive rate
plt.subplot(2, 1, 2)
plt.plot(df_grouped['date'], df_grouped['positive_rate'], label='Positive Rate', color='orange')
plt.title('Positive Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Positive Rate')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

#ACTIVITY 7 Visualize the fatality rate (total deaths / total cases) over time globally.

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('df_cleaned.csv')

data['date'] = pd.to_datetime(data['date'])
data['fatality_rate'] = data['total_deaths'] / data['total_cases']

global_fatality_rate = data.groupby('date')['fatality_rate'].mean().reset_index()
```
